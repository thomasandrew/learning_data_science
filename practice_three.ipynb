```
sets_by_year = sets.groupby('year').count()

plt.title('Lego', color='orange')
plt.ylabel('Number of Parts', fontsize=12, color='green')
plt.xlabel('Year', fontsize=12, color='blue')
plt.yticks(fontsize=10)
plt.xticks(fontsize=10)
plt.scatter(sets_by_year.index[:-2], sets_by_year.num_parts[:-2], color='red', linewidth=3, label='Num_parts')
plt.legend()
```

```
sets_by_year.rename(columns = {'theme_id':'nr_theme'}, inplace=True)
sets_by_year.head()
```

```
themes_by_year = sets.groupby('year').agg({'theme_id':pd.Series.nunique})

ax1 = plt.gca()
ax2 = ax1.twinx()

ax1.scatter(sets_by_year.index[:-2], sets_by_year.num_parts[:-2], color='green')
ax2.scatter(themes_by_year.index[:-2], themes_by_year.theme_id[:-2], color='red')

plt.title('Lego', color='orange')
ax1.set_ylabel('Number of Parts', color='green')
ax2.set_ylabel('Number of Themes', color='red')
plt.xlabel('Year', color='blue')
```

```
num_parts_by_year = sets.groupby('year').agg({'num_parts':pd.Series.nunique})

ax_one = plt.gca()
ax_two = ax_one.twinx()

ax_one.plot(new_sets.index[:-2], new_sets.set_num[:-2], label='Set_num', color='blue')
ax_two.plot(num_parts_by_year.index[:-2], num_parts_by_year.num_parts[:-2], color='green', label='Num_parts')

plt.title('Lego', color='orange')
ax_one.set_ylabel('Sets Numbers', color='blue')
ax_two.set_ylabel('Number of Parts', color='green')
plt.xlabel('Year', color='red')
plt.legend()
```
