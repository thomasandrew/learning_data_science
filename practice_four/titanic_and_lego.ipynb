# LEGO data

```
sets = pd.read_csv('sets.csv')

sets_by_numParts = sets.groupby('year').agg({'num_parts':pd.Series.nunique})

sets.head()
```

```
plt.plot(sets_by_numParts.index[:-2], sets_by_numParts.num_parts[:-2])
```

```
sets.head()
```

```
sets_by_setNum = sets.groupby('year').agg({'set_num': pd.Series.nunique})

x1 = plt.gca()
x2 = x1.twinx()

x1.plot(sets_by_numParts.index[:-2], sets_by_numParts.num_parts[:-2], color='red')
x2.plot(sets_by_setNum.index[:-2], sets_by_setNum.set_num[:-2], color='green')

plt.title('Lego', color='orange', fontsize=16)
x1.set_ylabel('Number of parts', color='red', fontsize=11)
x2.set_ylabel('Set of Numbers', color='green', fontsize=11)
```

```
sets.head()
```

```
sets_by_id = sets.groupby('year').agg({'theme_id' : pd.Series.mean})
plt.scatter(sets_by_id.index, sets_by_id.theme_id, label='Theme ID', linewidth=3, color='orange')
plt.legend()
```

```
class Test:
  def __init__(self, numSet, numParts, year, theme_id):
    self.numSet = numSet
    self.numParts = numParts
    self.year = year
    self.theme_id = theme_id

  def testFunc(self):
      print(plt.scatter(self.numSet.index[:-2], self.numSet.set_num[:-2], color='green', linewidth=3))
      print(plt.ylabel('Number of sets', color='Green', fontsize=14))
      print(plt.xlabel('Year', color='red', fontsize=14))

year = sets['year']

test = Test(sets_by_setNum, sets_by_numParts, year, sets_by_id)
test.testFunc()
```
