```
import pandas as pd
import matplotlib.pyplot as plt

dic = {'Linguagem de programacao': ["Java", "Javascript", "CPP", "Python", "C_sharp", "C"],
         'Ano': [1700, 1800, 1900, 2000, 2010,2020],
         'Post': [8.0,	85.0,	511.0,	164.0,	14.0,	0.0]
       }

dict = pd.DataFrame.from_dict(dic)
dict.head()
```

```
new_dict = dict.pivot(index="Ano", columns="Linguagem de programacao", values="Post")
new_dict.head()
```

```
new_dict.fillna(0, inplace=True)
new_dict.head()
```

```
#plt.figure(figsize=(5,4))
plt.title('Quantida de postagens em Prog em cada ano')
plt.yticks(fontsize=14)
plt.xticks(fontsize=14)
plt.ylabel('Post', fontsize=14)
plt.xlabel('Ano', fontsize=14)
#plt.plot(new_dict.index, new_dict['Java'])

for column in new_dict:
  plt.plot(new_dict.index, new_dict[column], linewidth=3, label=new_dict[column].name)
  plt.legend(fontsize=10)
```

```
class Exercicio:
    def __init__(self, c, cpp, c_sharp, java, javascript, python):
      self.c = c
      self.cpp = cpp
      self.c_sharp = c_sharp
      self.java = java
      self.javascript = javascript
      self.python = python

    def comparacao(self, ano):
      if ano == 1900 and self.cpp == 511.0:
        print(new_dict['CPP'].iloc[2])
      elif ano == 1700 and self.c == 0.0:
        print(new_dict['C'].iloc[0])
      elif ano == 2010 and self.c_sharp == 14.0:
        print(new_dict['C_sharp'].iloc[4])
      elif ano == 1700 and self.java == 8.0:
        print(new_dict['Java'].iloc[0])
      elif ano == 1800 and self.javascript == 85.0:
        print(new_dict['Javascript'].iloc[1])
      elif ano == 200 and self.python == 164.0:
        print(new_dict['Python'].iloc[3])
      else:
        print(False)

ano = 2010

exercicio = Exercicio(new_dict['C'].iloc[0], new_dict['CPP'].iloc[2], new_dict['C_sharp'].iloc[4], new_dict['Java'].iloc[0], new_dict['Javascript'].iloc[1], new_dict['Python'].iloc[3])
exercicio.comparacao(ano)
```
