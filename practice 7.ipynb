```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
import seaborn as sns
import numpy as np
from PIL import Image
```

```python
data = pd.read_csv('pokemon.csv')
```

```python
data.head()
```

```python
data.isna().any().sum()
```

```python
data.dropna(inplace=True)
```

```python
data.head()
```

```python
data.isna().any().sum()
```

```python
data.head()
```

```python
both = data.groupby('Name').agg({'Legendary' : pd.Series.bool})
both
```

```python
sns.catplot(x='Legendary', kind='count', hue='Legendary', palette='magma', data=both, height=4)

plt.xticks(rotation=45)
plt.title('Legendary Pokemon')
```
